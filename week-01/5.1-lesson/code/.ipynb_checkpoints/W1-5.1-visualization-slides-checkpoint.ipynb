# 5.1: Introduction to python visualization packages

Visualizing your data is **extremely** important to be proficient at as a data scientist. Why?

1. You need to be able to explore your data **visually**. This is essential!
    - You need to get an intuition for your data.
    - Visualization should **always** be done before you start modeling your data.
    - If you model data without visualizing it first, you are asking to run into problems down the line!

2. You will be always be required to report on your findings working as a data scientist. 
    - Technical coworkers such as other data scientists or analysts will want to get an intuition for the data too
    - Visualization will make your findings compelling and intuitive to non-technical coworkers.

### M.1 Plot out the years in job vs. the yearly salary as a scatter plot

````python
years_in_job = salary.years_in_job.values
yearly_salary = salary.yearly_salary.values
````

```plt.scatter()``` is the matplotlib function that will create a scatter plot. The x-axis data comes first, then the y-axis data.

- The x-axis variable comes first, then the y-axis variable.
- The c keyword argument is for the color of the data.

````python
### M.1
plt.scatter(years_in_job, yearly_salary, c='darkred')
````

### M.2: Make the figure bigger

````python
### M.2
plt.figure(figsize=(12,8))
plt.scatter(years_in_job, yearly_salary, c='darkred')
````

### M.3: Add a title and axis labels

Matplotlib charts do not come with labels attached, you have to specify them. The code changes and it's a bit tricky to understand:

- You need to return the figure object from the plt.figure() function.
- Pull out the current axes from the figure object with fig.gca() ["get current axes"].
- Call the scatter function with the axes object instead of the figure.
- Use the set_title(), set_xlabel(), and set_ylabe() functions to add the text.


```python
fig = plt.figure(figsize=(12,8))

axes = fig.gca()

axes.scatter(years_in_job, yearly_salary, c='darkred')

axes.set_title('years in job vs. yearly salary\n', fontsize=20)
axes.set_xlabel('years in job', fontsize=16)
axes.set_ylabel('yearly salary', fontsize=16)
```

````python
### M.3
fig = plt.figure(figsize=(12,8))
axes = fig.gca()
axes.scatter(years_in_job, yearly_salary, c='darkred')
axes.set_title('years in job vs. yearly salary\n', fontsize=20)
axes.set_xlabel('years in job', fontsize=16)
axes.set_ylabel('yearly salary', fontsize=16)
````

### M.4: Make the points a bit bigger and change the style

The points will be nicer bigger. We can change the style to look like Nate Silver's fivethirtyeight blog, which is one of the built in styles in matplotlib.

- The style is set with plt.style.use() first

### M.5: Plot a 3rd and 4th dimension using color and shape of the points. Add a legend

Shape and color can plot additional "dimensions" of the data. This is a great example of why visualizations are so important for understanding data. 

First we need to pull out the data for the new dimensions. Pull out gender and professor rank from the data.

````python
gender = salary.gender.values
rank = salary.professor_rank.values
````

The code for this is all of a sudden going to get totally insane...

It's written out for you in your notebook. Take a look at it, but don't expect to understand it right now. It's a good example of how matplotlib can be extended to do a lot with fine-grained control, but becomes very burdensome and very confusing.

```python
colors = ['darkred' if g == 'female' else 'steelblue' for g in gender]

# + == +, s == square, ^ == upwards triangle
point_shapes = ['+' if r == 'full' else 's' if r == 'associate' else '2' for r in rank]

plt.style.use('fivethirtyeight')
fig = plt.figure(figsize=(12,8))
axes = fig.gca()

for g in gender:
    
    if g == 'female':
        color = 'darkred'
    elif g == 'male':
        color = 'steelblue'
        
    for r in rank:
        
        if r == 'full':
            point_marker = '+'
        elif r == 'associate':
            point_marker = 's'
        elif r == 'assistant':
            point_marker = '^'
            
        current_years = [y for y, g_, r_ in zip(years_in_job, gender, rank) if g_ == g and r_ == r]
        current_salaries = [s for s, g_, r_ in zip(yearly_salary, gender, rank) if g_ == g and r_ == r]

        axes.scatter(current_years, current_salaries, c=color, marker=point_marker, s=150,
                     label=' '.join([r, g]))

axes.set_title('years in job vs. yearly salary\n', fontsize=20)
axes.set_xlabel('years in job', fontsize=16)
axes.set_ylabel('yearly salary', fontsize=16)

handles, labels = plt.gca().get_legend_handles_labels()
by_label = {l:h for l, h in zip(labels, handles)}
plt.legend(by_label.values(), by_label.keys(), loc = 'center right')
```

### S.2: Use the ```kind``` keyword argument to add a regression line

- The kind keyword argument can change the functionality of the plot.
- Setting this to "reg" will add a regression line with confidence intervals plotting the linear relationship between the two variables.

The "joint_kws" argument can be used to set matplotlib options under the hood, like the size of the points. This is shown below. Unfortunately this is not intuitive, but you can look up these answers on google as needed...

```python
sns.jointplot(years_in_job, yearly_salary, kind='reg', size=10, joint_kws={'scatter_kws': {'s': 150}})
```

### "pairplots" in seaborn

seaborn's "pairplot" is a really nice way of visualizing the final plot we did in matplotlib, but in a different way.

First, however, we are going to need to modify the data a bit...

### "wide" format vs "long" format [an aside]

We've been using data so far in something called **wide format**. There is another popular way of organizing data called **"long format**. What is long format?

Right now your columns of interest look like this:

    gender, rank, years, salary
    female, full, 12, 30000
    male, full, 18, 50000
    ...

But in long format, we only have **one column of numeric values.** It is going to look like this:

    gender, rank, variable, value
    female, full, years, 12
    female, full, salary, 30000
    male, full, years, 18
    male, full, salary, 50000
    ...

Every row corresponds to just one value, which is identified by the other columns.

The original format, which we will continue using for now, is called **wide** format. There are, as with everything, pros and cons to each.

Pandas has a nice convenience function for converting wide to long called melt:

````python
subset = salary[['gender','professor_rank','years_in_job','yearly_salary']]

subset_long = pd.melt(subset,
                      id_vars=['gender','professor_rank'],
                      value_vars=['years_in_job','yearly_salary'],
                      var_name='variable',
                      value_name='value')
````

### P.1: overview of the data objects

Plotly defines things to plot on a figure with specific objects. We need to do the different subsets of data separately.

Below is the code for the "male full" group:

```python
male_full = go.Scatter(
    x = years_in_job[(gender == 'male') & (rank == 'full')],
    y = yearly_salary[(gender == 'male') & (rank == 'full')],
    mode='markers',
    marker={'size':11,
            'line':{'width':0},
            'color':'steelblue',
            'symbol':"x",
            'opacity':0.65},
    name='Full Male'
)
```

### P.1: overview of the figure object

Once you have the data objects, you put them in a list and then define the figure object that holds them.

```python
data = [male_full, male_associate, male_assistant,
```
