Using the `staff_salary.csv` provided, write Python code to load the data and thereafter compute the following for the `SALARY` field:
* mean
* median

## Load data

```python
import csv
import numpy as np

nba_csv_path = 'staff_salary.csv'
rows = []

with open(nba_csv_path, 'r') as f:
    reader = csv.reader(f)
    for row in reader:
        rows.append(row)
```

## Assign header 

```python
header = rows[0]
data = rows[1:]
print(header)
```

```python
staff_data = {}

for index, column_name in enumerate(header):
    staff_data[column_name] = []
    for row in data:
        staff_data[column_name].append(row[index])
```

```python
staff_data['SALARY']
```

## Convert to float

```python
for name, col in staff_data.items():
    col = [float(x) for x in col]
    staff_data[name] = col
```

## Print mean and median salary

```python
print 'mean: ${:.2f}'.format(np.mean(staff_data['SALARY']))
print 'median: ${:.2f}'.format(np.median(staff_data['SALARY']))
```
