## Solutions

```python
list_one = np.array([1,3,5,7,9])
list_two = np.array([2,4,6,8,10])
```

```python
list_one
```

```python
list_two
```

```python
%%timeit
def dot_prod(a, b):
    total = 0
    for i in range(len(a)):
        mini_prod = a[i] * b[i]
        total += mini_prod
    return total
dot_prod(list_one, list_two)
```
