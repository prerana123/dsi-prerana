## As humans, we naturally model the world around ourselves

- We act upon beliefs we hold based upon past experience
- Most of the time our expectations match these mental models of the world
- When they don't we usually look foolish
- This gap between our beliefs and reality is the error
- If you have ever gone to pick up an emtpy container you thought was full, you have exeperienced this

## Example model

- When you drive at a constant speed s for t hours, the distance you travel is d = st.
- The quantities of distance and time are in a linear relationship because the graph is a line with slope (rate) s.
- This equation is a model between the distance travelled and the time travelled.

## Because we use rates to commonly understand natural processes, linear relationships are ubiquitous in our lives and our data. We don't always know how two quantities are related. Linear regression is a way to determine the relationship from a set of data.

- Models are relationships between quantities
- Linear regression is a method to determine the coefficients of linear relationships

# Prediction

## The benefit of having a model is to be able to make predictions

- In the example of picking up the carton, I can predict how much force I need to apply
- In the driving distance model, I can predict how far someone will drive
 - If they drive 2 hours at 60 mph, I can predict they will have driven 120 miles

## Predictions can be very valuable even if they are not always exactly right.

## "All models are wrong but some are useful" - George Box

> For example, if you own a movie theatre, you need need to make sure you have enough staff on nights and weekends to handle the increased number of patrons. A model to predict how many movie-goers will show up on a given night will help you predict the amount of employees that you'll need -- enough to handle the customers but not too many so that you keep labor costs reasonable. Too few employees and you won't be able to serve all the customers; too many and you've spent too much on labor. The better the estimate the better your business runs, and even if the estimate is off a bit you are still more prepared than if you had no model at all.

## Companies rely upon these models to effectively operate their businesses

## Examples:
> Credit Cards: Fraud Prevention

> Movie Studios: Ticket Sales

> Insurance: Claims Forecasting

# Linear regression

## What is linear regression? Simply put it is a method of determining the coefficients of a linear combination of variables:


$$ y = \alpha_0 x_0 + \alpha_0 x_1 + ... \alpha_n x_n$$

## To use linear regression, the variables themselves need not be linear, just the relationships between the quantities. So we could, for example, fit a target variable y to a polynomial:

$$ y = \alpha_0 + \alpha_1 x_1 + ... \alpha_2 x_2{^2}$$

## In this case we will have to supply <br><br>both x and x<sup>2</sup> for every data point. 

## For example, the trajectory of a ball thrown through the air forms a parabola with equation:

$$ d = d_0 + v t + 0.5 a t^2$$ 

Where d is the distance travelled, v is the velocity, and a is the acceleration due to gravity (9.8 m / s<sup>2</sup> on the surface of Earth). <br>We could determine the coefficients v and a from a set of data using linear regression.

## Eyeballing a Linear Regression

<center>Equation of a line:</center>
$$ y = mx + b $$

<center>Finding the slope:</center>
$$ m = (y_1 - y_0) / (x_1 - x_0) $$ 

<center>Point slope formula:</center>
$$ y - y_0 = m (x - x_0) $$
