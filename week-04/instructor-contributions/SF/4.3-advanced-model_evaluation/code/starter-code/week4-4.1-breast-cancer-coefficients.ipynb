## Interpreting Pearson

* The correlation coefficient ranges from −1 to 1.
* A value of 1 implies that a linear equation describes the relationship between X and Y perfectly, with all data points lying on a line for which Y increases as X increases. 
* A value of −1 implies that all data points lie on a line for which Y decreases as X increases. 
* A value of 0 implies that there is no linear correlation between the variables.

#### Another way to do a pearson coeficient matrix is using the built-in Pandas df.corr() method

```python
coefs = df[df.columns[2:12]].corr()
coefs
```

#### Using this in conjunction with a heatmap can give a quick overview of anything interesting.

```python
%matplotlib inline
```

```python
import seaborn as sns
import matplotlib, matplotlib.pyplot as plt

# initializing subplots allows changing of font size
fig, ax = plt.subplots()
fig.set_size_inches(12, 8)

# Change font size
matplotlib.rcParams.update({'font.size': 12})

# Sets x/y label sizes
cax = plt.gcf().axes[-1]
cax.tick_params(labelsize=15)

plot = sns.heatmap(coefs, annot=True, ax=ax)
plot.set_title("Correlation Heatmap")
plt.xticks(rotation=50) 
# plot.set_xticklabels(rotation=30)
```
