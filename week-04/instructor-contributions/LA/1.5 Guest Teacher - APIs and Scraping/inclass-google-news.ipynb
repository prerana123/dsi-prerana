```python
import requests
from bs4 import BeautifulSoup
```

```python
GOOGLE_URL = 'http://news.google.com'
r = requests.get(GOOGLE_URL)
```

```python
bs = BeautifulSoup(r.text, "lxml")
```

```python
# Gets all span tags with class 'titletext'
tags = bs.select('span.titletext')

for tag in tags:
    print('+', tag.text)
```

```python
# Gets all links on the page
tags = bs.select('a')

for tag in tags:
    if 'href' in tag.attrs:
        print('+', tag.text, tag.attrs['href'])
```
