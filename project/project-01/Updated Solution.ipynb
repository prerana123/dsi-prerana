##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
def sd1(clm):
    val = [(i-numpy.mean(clm))**2 for i in clm]
    sum1 =numpy.sum(val)
    va = sum1/len(clm)
    return(va **.5)

sd1(dic1['Rate'])
sd1(dic1['Verbal'])
sd1(dic1['Math'])
print("Standard Deviation of Rate clm is", sd1(dic1['Rate']))
print("Standard Deviation of Math clm is", sd1(dic1['Math']))  
print("Standard Deviation of Verbal clm is", sd1(dic1['Verbal']))
#numpy.std(dic1['Rate'])
print(numpy.mean(dic1['Rate']))#mean rate
print(numpy.mean(dic1['Math']))
print(numpy.mean(dic1['Verbal']))
print(numpy.median(dic1['Rate']))#medain rate
print(numpy.median(dic1['Math']))
print(numpy.median(dic1['Verbal']))
print(numpy.var(dic1['Verbal']))
print(numpy.var(dic1['Math']))
```
