## Step 2: Clean data

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that are too long, and check for missing values (replace them and explain rationale if it makes sense).

```python
#Drop column with all null value from Data Frames
bd1 = bd.dropna(axis=1,how = "all")

```

```python
# to verify all null value column .
#bd.isnull()
```

```python
#Columns "date.entered" and "date.peaked" are converted to datetime for ease of calculation
#and Time is converted to string

bd1.loc[:,"date.entered"] = pd.to_datetime(bd1.loc[:,"date.entered"])
bd1.loc[:,"date.peaked"] = pd.to_datetime(bd1.loc[:,"date.peaked"])
bd1.loc[:,"time"] = bd1.loc[:,"time"].astype(str)


bd1.head(3)

```

```python
#to check column with new data type.
bd1.dtypes
```

```python
# Code to convert time column to float for furthur calculations

list2 = []
list3 = []
for i in bd1["time"]:
    v1 = i
    v2 = v1.replace(":",".")
    list2.append(v2)

```

```python
# Code to convert time column to float for furthur calculations 
list3 = list2 
list3 = map(float,list3)
bd1.loc[:,"time"] = list3
```

##### Using Pandas' built in `melt` function, pivot the weekly Scoreing data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Scoreing. There will now be multiple entries for each song, one for each week on the Billboard Scoreings.

```python
# Week columns were converted to rows and was assigned a Variable value using melt function
# this will make data more readable and ready for furthur processing
bd2 =pd.melt(bd1,id_vars = ['artist.inverted','track','genre','time', 'date.entered', 'date.peaked'],value_vars = list(bd1.iloc[:,7:]))

bd2.head(4)
```

```python
#Added Days to Peak Column by substracting Date Peaked from Date Entered
bd2.loc[:,"days_to_peak"] = bd2.loc[:,"date.peaked"]-bd2.loc[:,"date.entered"]
bd2.head(4)
```

```python
#Value Variable="x1st.week...etc" is good to read but is not good to perform any calculations
#on varibale column converting this to No of Weeks which will be useful later on for 
#plotting weekly graphs
list1= []
for i in bd2["variable"]:
    v1 = i
    v2 = v1.replace("x","").replace("th","").replace("st","").replace("rd","").replace("nd","").replace(".week","")         
    list1.append(v2)
list1 = map(int,list1)  
```

```python
bd2["variable"] = list1
bd2.head(2)
bd2.shape
```

As matter of policy, Billboard removes former hits from the Hot 100 once they drop below 50. Billboard’s reasoning is that it gives new songs and new artists a better chance at getting on the charts. That’s probably a good thing from a music industry reporting perspective, but it’s not good from our data science perspective.Perhaps the best approach is to simply only pay attention to Scoreings from 1- 50. 

```python
#Messaged the Data to Include only top 50 ratings as that will make more sence due 
#to above mentioned Billboard policy
#Also Included a Score Column by a calculating 51 - Value(Rating) this no will be 
#later on Score will be summed up to calculate various results   
#Higer the Score more popular the track,artist or genre
bd3 = bd2[bd2["value"]<50]
bd3.loc[:,"value"] = bd3.loc[:,"value"].astype(int)
bd3.loc[:,"Score"] = 51-bd3.loc[:,"value"]
bd3.head(4)
```

```python
#Renamed Variable column to week ,value to rating etc...
bd4 = bd3.rename(columns = {"variable":"week","value": "rating","genre":"genre","artist.inverted":"artist"})
bd4.head(10)
```
