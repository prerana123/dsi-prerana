## Part 3: Data Wrangling

```python
# making dummy for Embarked and Sex column
```

```python
dummies_Embarked = pd.get_dummies(df3.Embarked)
dummies_Sex = pd.get_dummies(df3.Sex)

df4 = pd.concat([df3, dummies_Embarked,dummies_Sex], axis=1)
df4.head(2)
```

```python

df4 = df4.rename(columns={'C': 'Embarked_C', 'Q': 'Embarked_Q','S':'Embarked_S'})
```

```python
df5 = df4.drop(["Embarked","Embarked_S","Sex","male"],axis=1) # drop the column
df5.head(2)
```

```python
df5.columns
```

```python
X = df5[["Pclass","Age","SibSp","Parch","Fare","Embarked_C","Embarked_Q","female"]]
column = ["Pclass","Age","SibSp","Parch","Fare","Embarked_C","Embarked_Q","female"]
Y = df5["Survived"]
df5.head(2)
```

## Part 4: Logistic Regression and Model Validation

```python
#scaling 
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X = scale.fit_transform(X)
X[0:2]
```

```python
df6 = pd.DataFrame(X,columns = column )
df6[0:2]
```

```python
# fit model
logreg = LogisticRegression()
logreg.fit(X, Y)
```

```python
logreg.score(X,Y)
```

```python
pre_y = logreg.predict(X)

print(pre_y[0:5],Y[0:5])
logreg.predict_proba(X)
```

```python
# examine coefficients
coeffs = pd.DataFrame(logreg.coef_, columns = column)
coeffs
coeft =  coeffs.transpose()
coeft.columns = ["logistic coeffs"]
coeft
```

```python
# DL. This is a great representation of your coefs!
# You should spend some time trying to intepret you coefs in terms of odds/likelihood.
coeft.sort_values("logistic coeffs").plot(kind = "bar")
```

```python
#more women are survived than man.
```

```python
# split and test generalization
from sklearn.cross_validation import cross_val_score
from sklearn.cross_validation import cross_val_score, train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33,random_state=50)
logreg_cv =  LogisticRegressionCV(cv=5,Cs=5,random_state=50)
logreg_cv.fit(X_train,Y_train)
Y_pred = logreg_cv.predict(X_test)
ypro = logreg_cv.predict_proba(X_test)
Y_score = logreg_cv.decision_function(X_test)

cv_score = cross_val_score(logreg, X, Y)

```

```python
cv_score
```

```python
mean = cv_score.mean()
std = cv_score.std()
print(mean,std)
```

```python
logreg_cv.score(X_test,Y_test)
```

#### What does the Confusion Matrix tell us? 

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
conmat = np.array(confusion_matrix(Y_test, Y_pred))
confusion = pd.DataFrame(conmat, index=['Survived', 'Not Survived'],
                         columns=['predicted_Survived', 'predicted_NotSurvived'])
confusion
```

```python
TP = confusion.ix['Survived', 'predicted_Survived']
FP = confusion.ix['Not Survived', 'predicted_Survived']
TN = confusion.ix['Not Survived', 'predicted_NotSurvived']
FN = confusion.ix['Survived', 'predicted_NotSurvived']
print(zip(['True Positives','False Positives','True Negatives','False Negatives'],
         [TP, FP, TN, FN]))
```

```python
# compute accuracy
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
acc = accuracy_score(Y_test, Y_pred)
print "Accuracy" , acc
```

## Part 6: Gridsearch and kNN
Perform Gridsearch for the same classification problem as above, but use KNeighborsClassifier as your estimator, and number of neighbors and weights as test parameters.

```python
knc = KNeighborsClassifier()

knc_param = {'n_neighbors': range(3, 10),
             'weights': ['uniform', 'distance']}

grid = GridSearchCV(knc,knc_param, cv=5)
grid = GridSearchCV(knc,knc_param, cv=5)
grid.fit(X, Y)

print grid.best_params_
print grid.best_score_
```

```python
kn_score = cross_val_score(grid.best_estimator_,X,Y)
print(kn_score.mean(),kn_score.std())
```

```python
# new kNN model with the optimal parameters
bestKNN = grid.best_estimator_
bestKNN.fit(X_train, Y_train)
knnpred_y = bestKNN.predict(X_test)
print cross_val_score(bestKNN, X, Y).mean()
print cross_val_score(bestKNN, X, Y).std()
```

```python
# new confusion matrix; what has changed?
from sklearn.metrics import confusion_matrix
conmat = np.array(confusion_matrix(Y_test, knnpred_y))
confusion = pd.DataFrame(conmat, index=['Survived', 'Not Survived'],
                         columns=['predicted_Survived', 'predicted_NotSurvived'])
confusion
```
